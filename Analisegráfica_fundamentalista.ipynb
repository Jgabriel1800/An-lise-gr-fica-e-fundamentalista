```python
import pandas as pd
carteira_fund = ['WEGE3','ABEV3','GGBR4','ELET3','ITSA4','B3SA3', 'PETR4','SUZB3','RENT3','VALE3']

lista_dfs = []
for papel in carteira_fund:
    try:

        df = fundamentus.get_papel(papel)
        df = df[['Setor','Cotacao','Min_52_sem','Max_52_sem','Valor_de_mercado','Nro_Acoes','Patrim_Liq','Receita_Liquida_12m','Receita_Liquida_3m','Lucro_Liquido_12m','Lucro_Liquido_3m']]

        lista_dfs.append(df)
    except Exception as e:
        print(f"Erro ao obter dados para {papel}: {e}")

ind = pd.concat(lista_dfs, keys=carteira_fund)

ind = ind.reset_index(level=0)

ind.rename(columns={"level_0":"Ativo"},inplace=True)
ind=ind.reset_index(drop=True)

ind.head(3)
```

```python
ind.info()
```

Alterando colunas de object para numerico

```python
colunas=['Cotacao','Min_52_sem','Max_52_sem','Valor_de_mercado','Nro_Acoes','Patrim_Liq','Receita_Liquida_12m','Receita_Liquida_3m','Lucro_Liquido_12m','Lucro_Liquido_3m']

ind[colunas]=ind[colunas].apply(pd.to_numeric,errors='coerce',axis=1)
ind.head()
```

```python
ind.info()
```

Criand o um com indicadores da carteira

```python
ind_2=fundamentus.get_resultado_raw().reset_index()

ind_2=ind_2.query("papel in @carteira_fund")

ind_2=ind_2[['papel','P/L','Div.Yield','P/VP','ROE']].reset_index(drop=True)

ind_2.rename(columns={'papel':'Ativo','Div.Yield':'DY'},inplace=True)

ind_2.head()
```

concatenação de df

```python
indicadores=pd.merge(ind,ind_2,on='Ativo')

indicadores['LPA']=(indicadores['Lucro_Liquido_12m']/indicadores['Nro_Acoes']).round(2)

indicadores['VPA']=(indicadores['Patrim_Liq']/indicadores['Nro_Acoes']).round(2)


indicadores.head()
```
